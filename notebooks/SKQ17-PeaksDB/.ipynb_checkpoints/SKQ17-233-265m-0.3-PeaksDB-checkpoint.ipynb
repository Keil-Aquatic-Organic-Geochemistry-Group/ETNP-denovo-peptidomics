### Manipulation of PeaskDB _de novo_-assisted database search results of ETNP 2017 P2 samples LC-MS/MS data using python.

Starting with:

    PeaksDB search results (.csv) of PTM-optimized database searches
    These were all searched with 15 ppm precursor tolerance and 0.5 ppm fragement ion tolerance
    Search database included marine fungi and labrinthulomyces discovered using _de novo_ peptide sequencing
    XInteract file includes precursor intensities and protein descriptions

Goal:

    Files with stripped (no PTMs) peptide lists and
    Columns with #'s of each modification in every sequence
    Column with stripped peptide lengths (# amino acids)
    
### To use for a different file:

#### 1. Change the input file name in *IN 4*
#### 2. Use 'find + replace' (Esc + F) to replace the running # (e.g., 233) for another
#### 3. Update the NAAF factor calculated in *IN 6* into *IN 7*

## 233: 265 m McLane pump filtered on 0.3 um GF-75

```python
# read the CSV into a dataframe using the pandas read_csv function
peaksdbdup233 = pd.read_csv("/home/millieginty/Documents/git-repos/2017-etnp/data/pro2020/ETNP-SKQ17/PeaksDB-PTMopt/233/ETNP_SKQ17_PEAKSDB_163_233-265m-0.3-JA4_15ppm_peptide.csv")

# remove redundant rows
peaksdb233 = pd.DataFrame.drop_duplicates(peaksdbdup233)

print(peaksdb233.columns)

columns = ['Peptide', '-10lgP', 'Mass', 'Length', 'ppm', 'm/z', 'RT',
       'Area', 'Fraction', 'Scan', 'Source File',
       '#Spec', '#Spec', 'Accession', 'PTM',
       'AScore']

peaksdb233.columns = columns

print("# redundant peaksdb peptides in combined dataframe", len(peaksdbdup233))
print("# nonredundant peaksdb peptides in combined dataframe", len(peaksdb233))

#look at the dataframe
peaksdb233.head()
```

```python
# use a count function to enumerate the # of A's (alanines) in each peptide
peaksdb233['A'] = peaksdb233['Peptide'].str.count("A")

# use a count function to enumerate the # of C's (cysteines) in each peptide
peaksdb233['C'] = peaksdb233['Peptide'].str.count("C")

# use a count function to enumerate the # of D's (aspartic acids) in each peptide
peaksdb233['D'] = peaksdb233['Peptide'].str.count("D")

# use a count function to enumerate the # of E's (glutamic acids) in each peptide
peaksdb233['E'] = peaksdb233['Peptide'].str.count("E")

# use a count function to enumerate the # of F's (phenylalanines) in each peptide
peaksdb233['F'] = peaksdb233['Peptide'].str.count("F")

# use a count function to enumerate the # of G's (glycines) in each peptide
peaksdb233['G'] = peaksdb233['Peptide'].str.count("G")

# use a count function to enumerate the # of H's (histidines) in each peptide
peaksdb233['H'] = peaksdb233['Peptide'].str.count("H")

# use a count function to enumerate the # of I's (isoleucines) in each peptide
# in peaksdb233 output, there will be no isoleucines (they're lumped in with leucines)
peaksdb233['I'] = peaksdb233['Peptide'].str.count("I")

# use a count function to enumerate the # of K's (lysines) in each peptide
peaksdb233['K'] = peaksdb233['Peptide'].str.count("K")

# use a count function to enumerate the # of L's (leucines) in each peptide
# also these include the isoleucines
peaksdb233['L'] = peaksdb233['Peptide'].str.count("L")

# use a count function to enumerate the # of M's (methionines) in each peptide
peaksdb233['M'] = peaksdb233['Peptide'].str.count("M")

# use a count function to enumerate the # of N's (asparagines) in each peptide
peaksdb233['N'] = peaksdb233['Peptide'].str.count("N")

# use a count function to enumerate the # of P's ([prolines]) in each peptide
peaksdb233['P'] = peaksdb233['Peptide'].str.count("P")

# use a count function to enumerate the # of Q's (glutamines) in each peptide
peaksdb233['Q'] = peaksdb233['Peptide'].str.count("Q")

# use a count function to enumerate the # of R's (arginines) in each peptide
peaksdb233['R'] = peaksdb233['Peptide'].str.count("R")

# use a count function to enumerate the # of S's (serines) in each peptide
peaksdb233['S'] = peaksdb233['Peptide'].str.count("S")

# use a count function to enumerate the # of T's (threonines) in each peptide
peaksdb233['T'] = peaksdb233['Peptide'].str.count("T")

# use a count function to enumerate the # of V's (valines) in each peptide
peaksdb233['V'] = peaksdb233['Peptide'].str.count("V")

# use a count function to enumerate the # of W's (tryptophans) in each peptide
peaksdb233['W'] = peaksdb233['Peptide'].str.count("W")

# use a count function to enumerate the # of Y's (tyrosines) in each peptide
peaksdb233['Y'] = peaksdb233['Peptide'].str.count("Y")

# use a count function to enumerate the # of carbamidomethylated C's in each peptide
peaksdb233['c-carb'] = peaksdb233['Peptide'].str.count("57.02")

# use a count function to enumerate the # of oxidized M's in each peptide
peaksdb233['m-oxid'] = peaksdb233['Peptide'].apply(lambda x: x.count('M(+15.99)'))

# use a lamba function to enumerate the # of deamidated N's in each peptide
peaksdb233['n-deam'] = peaksdb233['Peptide'].apply(lambda x: x.count('N(+.98)'))

# use a count function to enumerate the # of deamidated Q's in each peptide
peaksdb233['q-deam'] = peaksdb233['Peptide'].apply(lambda x: x.count('Q(+.98)'))

# use a count function to enumerate the # of hydroxylated K's in each peptide
peaksdb233['k-hydr'] = peaksdb233['Peptide'].apply(lambda x: x.count('K(+15.99)'))

# use a count function to enumerate the # of hydroxylated P's in each peptide
#peaksdb233['p-hydr'] = peaksdb233['Peptide'].apply(lambda x: x.count('P(+15.99)'))

# use a count function to enumerate the # of methylated R's in each peptide
peaksdb233['r-meth'] = peaksdb233['Peptide'].apply(lambda x: x.count('R(+14.02)'))

# create a column with 'stripped' peptide sequences using strip
peaksdb233['stripped peptide'] = peaksdb233['Peptide'].str.replace(r"\(.*\)","")

# add a column with the stripped peptide length (number of AAs)
peaksdb233['stripped length'] = peaksdb233['stripped peptide'].apply(len)

peaksdb233['NAAF num.'] = peaksdb233['Area'] / peaksdb233['stripped length']

# total the number of modifications in sequence
peaksdb233['ptm-total'] = peaksdb233['c-carb'] + peaksdb233['m-oxid'] + peaksdb233['n-deam'] + peaksdb233['q-deam'] + peaksdb233['k-hydr'] + peaksdb233['r-meth']

# write modified dataframe to new txt file
peaksdb233.to_csv("/home/millieginty/Documents/git-repos/2017-etnp/data/pro2020/ETNP-SKQ17/PeaksDB-PTMopt/233/ETNP_SKQ17_PEAKSDB20_233-265m-0.3-JA4_15ppm.csv")

# check out the results
peaksdb233.head()
```

```python
# made a new dataframe that contains the sums of certain columns in the modified
# peptide dataframe above 

index = ['sample total']

data = {'A': peaksdb233['A'].sum(),
        'C': peaksdb233['C'].sum(),
        'D': peaksdb233['D'].sum(),
        'E': peaksdb233['E'].sum(),
        'F': peaksdb233['F'].sum(),
        'G': peaksdb233['G'].sum(),
        'H': peaksdb233['H'].sum(),
        'I': peaksdb233['I'].sum(),
        'K': peaksdb233['K'].sum(),
        'L': peaksdb233['L'].sum(),
        'M': peaksdb233['M'].sum(),
        'N': peaksdb233['N'].sum(),
        'P': peaksdb233['P'].sum(),
        'Q': peaksdb233['Q'].sum(),
        'R': peaksdb233['R'].sum(),
        'S': peaksdb233['S'].sum(),
        'T': peaksdb233['T'].sum(),
        'V': peaksdb233['V'].sum(),
        'W': peaksdb233['W'].sum(),
        'Y': peaksdb233['Y'].sum(),
        'c-carb': peaksdb233['c-carb'].sum(),
        'm-oxid': peaksdb233['m-oxid'].sum(),
        'n-deam': peaksdb233['n-deam'].sum(),
        'q-deam': peaksdb233['q-deam'].sum(),
        'k-hydr': peaksdb233['k-hydr'].sum(),
        'r-meth': peaksdb233['r-meth'].sum(),
        'Total area': peaksdb233['Area'].sum(),
        'Total length': peaksdb233['stripped length'].sum()
       }

totalpeaksdb233 = pd.DataFrame(data, columns=['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', 'c-carb', 'm-oxid', 'n-deam', 'q-deam', 'k-hydr', 'r-meth', 'Total area', 'Total length'], index=index)

# calculate percentage of C's with carb (should be 1.0)
totalpeaksdb233['% C w/ carb'] = totalpeaksdb233['c-carb'] / totalpeaksdb233['C'] 

# calculate percentage of M's that are oxidized
totalpeaksdb233['% M w/ oxid'] = totalpeaksdb233['m-oxid'] / totalpeaksdb233['M'] 

# calculate percentage of N's that are deamidated
totalpeaksdb233['% N w/ deam'] = totalpeaksdb233['n-deam'] / totalpeaksdb233['N'] 

# calculate percentage of Q's that are deamidated
totalpeaksdb233['% Q w/ deam'] = totalpeaksdb233['q-deam'] / totalpeaksdb233['Q'] 

# calculate percentage of K's that are hydroxylated
totalpeaksdb233['% K w/ hydr'] = totalpeaksdb233['k-hydr'] / totalpeaksdb233['K'] 

# calculate percentage of P's that are hydroxylated
#totalpeaksdb233['% P w/ hydr'] = totalpeaksdb233['p-hydr'] / totalpeaksdb233['P'] 

# calculate percentage of R's that are methylated
totalpeaksdb233['% R w/ meth'] = totalpeaksdb233['r-meth'] / totalpeaksdb233['R'] 

# calculate NAAF denominator for all peptides in dataset i
totalpeaksdb233['NAAF denom.'] = totalpeaksdb233['Total area'] / totalpeaksdb233['Total length']

# write modified dataframe to new txt file
totalpeaksdb233.to_csv("/home/millieginty/Documents/git-repos/2017-etnp/data/pro2020/ETNP-SKQ17/PeaksDB-PTMopt/233/ETNP_SKQ17_PEAKSDB20_233-265m-0.3-JA4_15ppm_totals.csv")

totalpeaksdb233.head()
```

```python
# use the calculated NAAF factor (in totalpeaksdb233 dataframe, above) to caluclate the NAAF 
# NAAF: normalized normalized area abundance factor

NAAF20 = 213827.669181

# use NAAF >-10lgP to get NAAF
peaksdb233['NAAF factor'] = (peaksdb233['NAAF num.'])/NAAF20

# make a dataframe that contains only what we need: sequences, AAs, PTMs
peaksdb233_NAAF = peaksdb233[['stripped peptide', 'NAAF factor', 'A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', \
                                'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', 'c-carb', 'm-oxid', 'n-deam', \
                                'q-deam', 'k-hydr', 'r-meth']].copy()

# multiply the NAAF20 factor by the AAs to normalize its abundance by peak area and peptide length

peaksdb233_NAAF['A-NAAF20'] = peaksdb233_NAAF['A'] * peaksdb233['NAAF factor']
peaksdb233_NAAF['C-NAAF20'] = peaksdb233_NAAF['C'] * peaksdb233['NAAF factor']
peaksdb233_NAAF['D-NAAF20'] = peaksdb233_NAAF['D'] * peaksdb233['NAAF factor']
peaksdb233_NAAF['E-NAAF20'] = peaksdb233_NAAF['E'] * peaksdb233['NAAF factor']
peaksdb233_NAAF['F-NAAF20'] = peaksdb233_NAAF['F'] * peaksdb233['NAAF factor']
peaksdb233_NAAF['G-NAAF20'] = peaksdb233_NAAF['G'] * peaksdb233['NAAF factor']
peaksdb233_NAAF['H-NAAF20'] = peaksdb233_NAAF['H'] * peaksdb233['NAAF factor']
peaksdb233_NAAF['I-NAAF20'] = peaksdb233_NAAF['I'] * peaksdb233['NAAF factor']
peaksdb233_NAAF['K-NAAF20'] = peaksdb233_NAAF['K'] * peaksdb233['NAAF factor']
peaksdb233_NAAF['L-NAAF20'] = peaksdb233_NAAF['L'] * peaksdb233['NAAF factor']
peaksdb233_NAAF['M-NAAF20'] = peaksdb233_NAAF['M'] * peaksdb233['NAAF factor']
peaksdb233_NAAF['N-NAAF20'] = peaksdb233_NAAF['N'] * peaksdb233['NAAF factor']
peaksdb233_NAAF['P-NAAF20'] = peaksdb233_NAAF['P'] * peaksdb233['NAAF factor']
peaksdb233_NAAF['Q-NAAF20'] = peaksdb233_NAAF['Q'] * peaksdb233['NAAF factor']
peaksdb233_NAAF['R-NAAF20'] = peaksdb233_NAAF['R'] * peaksdb233['NAAF factor']
peaksdb233_NAAF['S-NAAF20'] = peaksdb233_NAAF['S'] * peaksdb233['NAAF factor']
peaksdb233_NAAF['T-NAAF20'] = peaksdb233_NAAF['T'] * peaksdb233['NAAF factor']
peaksdb233_NAAF['V-NAAF20'] = peaksdb233_NAAF['V'] * peaksdb233['NAAF factor']
peaksdb233_NAAF['W-NAAF20'] = peaksdb233_NAAF['W'] * peaksdb233['NAAF factor']
peaksdb233_NAAF['Y-NAAF20'] = peaksdb233_NAAF['Y'] * peaksdb233['NAAF factor']

# multiply the NAAF20 factor by the PTMs normalize its abundance by peak area and peptide length

peaksdb233_NAAF['ccarb-NAAF20'] = peaksdb233_NAAF['c-carb'] * peaksdb233_NAAF['NAAF factor']
peaksdb233_NAAF['moxid-NAAF20'] = peaksdb233_NAAF['m-oxid'] * peaksdb233_NAAF['NAAF factor']
peaksdb233_NAAF['ndeam-NAAF20'] = peaksdb233_NAAF['n-deam'] * peaksdb233_NAAF['NAAF factor']
peaksdb233_NAAF['qdeam-NAAF20'] = peaksdb233_NAAF['q-deam'] * peaksdb233_NAAF['NAAF factor']
peaksdb233_NAAF['khydr-NAAF20'] = peaksdb233_NAAF['k-hydr'] * peaksdb233_NAAF['NAAF factor']
peaksdb233_NAAF['rmeth-NAAF20'] = peaksdb233_NAAF['r-meth'] * peaksdb233_NAAF['NAAF factor']

# write the dataframe to a new csv
peaksdb233_NAAF.to_csv("/home/millieginty/Documents/git-repos/2017-etnp/data/pro2020/ETNP-SKQ17/PeaksDB-PTMopt/233/ETNP_SKQ17_PEAKSDB20_233-265m-0.3-JA4_15ppm_NAAF.csv")

peaksdb233_NAAF.head()
```

### Using BioPython to query peptide sequences

I installed the BioPython package using `pip install biopython`. All instructions and information [here](https://www.tutorialspoint.com/biopython/index.htm). 

GitHub project: https://github.com/biopython/biopython

I'm relying on the ProtParam module to parse sequences for relative AA composition, instability, secondary structure, instability, and hydrophobicity. You can read more about that module and the studies the indecies are derived from here:

https://biopython.org/wiki/ProtParam

```python
# Bio.SeqIO is the standard Sequence Input/Output interface for BioPython 1.43 and later
# Bio.SeqIO provides a simple uniform interface to input and output assorted sequence file formats.
# (including multiple sequence alignments), but will only deal with sequences as SeqRecord objects

# for accepted file formats see https://biopython.org/wiki/SeqIO

from Bio import SeqIO
#for seq_record in SeqIO.parse("/home/millieginty/Documents/git-repos/2017-etnp/data/MED4/MED2_tryp_1raw_db_peptides_nmod.fasta", "fasta"):
    #print(seq_record.id)
    #print(repr(seq_record.seq))
    #print(len(seq_record))
    
# I commented the print functions out so the output doesn't take up too much space. 
```

```python
# seeing what the ProtParam module can do with a single protein sequence:

from Bio.SeqUtils.ProtParam import ProteinAnalysis

test_seq = "MAEGEITTFTALTEKFNLPPGNYKKPKLLYCSNGGHFLRILPDGTVDGTRDRSDQHIQLQLSAESVGEVYIKSTETGQYLAMDTSGLLYGSQTPSEECLFLERLEENHYNTYTSKKHAEKNWFVGLKKNGSCKRGPRTHYGQKAILFLPLPV"

analysed_seq = ProteinAnalysis(test_seq)
print("molecular weight of seq =", analysed_seq.molecular_weight())

# calculates the aromaticity value of a protein according to Lobry & Gautier (1994, Nucleic Acids Res., 22, 3174-3180). 
# it's simply the relative frequency of Phe+Trp+Tyr.

analysed_seq.aromaticity()
print("aromaticity of seq =", analysed_seq.aromaticity())

# secondary_structure_fraction:
# this methods returns a list of the fraction of amino acids which tend to be in helix, turn or sheet. 
# AAs in helix: V, I, Y, F, W, L
# AAs in turn: N, P, G, S
# AAs in sheet: E, M, A, L
# the returned list contains 3 values: [Helix, Turn, Sheet]

analysed_seq.secondary_structure_fraction()
print("frac in H T S =", analysed_seq.secondary_structure_fraction())

# the instability index, an implementation of the method of Guruprasad et al. (1990, Protein Engineering, 4, 155-161).
# this method tests a protein for stability. 
# any value above 40 means the protein is unstable (=has a short half life)
# NOT SURE WHAT THIS MEANS FOR PEPTIDES, BUT WE COULD DO THIS FOR PROTEINS

analysed_seq.instability_index()
print("instability =", analysed_seq.instability_index())

# count_amino_acids will do just that, and get_amino_acids_percent will return %'s for each AA across the sequence. 
analysed_seq.get_amino_acids_percent()

# taking the returned dictionary and converting to a dataframe

aadict = analysed_seq.get_amino_acids_percent()
aadf = pd.DataFrame(list(aadict.items()),columns = ['residue','% occurance']) 

aadf.head()
```

```python
# use SeqIO and a loop to apply count_amino_acids to each sequence in the file
# aatot will give us the total number of each residue in the entire sample output

import collections
from Bio import SeqIO
from Bio.SeqUtils.ProtParam import ProteinAnalysis

all_aas = collections.defaultdict(int)
for record in SeqIO.parse("/home/millieginty/Documents/git-repos/2017-etnp/data/pro2020/ETNP-SKQ17/PEAKS-PTMopt/ETNP-SKQ17-231-100m-0.3-JA2_DN50_stripped_peptides.fas", "fasta"):
    x = ProteinAnalysis(str(record.seq))
    #print(record.id, x.count_amino_acids())
    for aa, count in x.count_amino_acids().items():
        all_aas[aa] += count        
        
# made a dataframe for amino acid total counts        
data = (all_aas)
aatot = pd.DataFrame(data, index = ['sample sequence total'])
aatot.head()
```

```python
from pandas import Series, DataFrame

with open('/home/millieginty/Documents/git-repos/2017-etnp/data/pro2020/ETNP-SKQ17/PEAKS-PTMopt/ETNP-SKQ17-231-100m-0.3-JA2_DN50_stripped_peptides.fas') as fasta_file:  # Will close handle cleanly
    identifiers = []
    lengths = []
    for seq_record in SeqIO.parse(fasta_file, 'fasta'):  # (generator)
        identifiers.append(seq_record.id)
        lengths.append(len(seq_record.seq))
        
        
#converting lists to pandas Series    
s1 = Series(identifiers, name='ID')
s2 = Series(lengths, name='length')

#Gathering Series into a pandas DataFrame and rename index as ID column
idseq = DataFrame(dict(ID=s1, length=s2)).set_index(['ID'])

idseq.head()
```

```python
from pandas import Series, DataFrame



with open('/home/millieginty/Documents/git-repos/2017-etnp/data/pro2020/ETNP-SKQ17/PEAKS-PTMopt/ETNP-SKQ17-231-100m-0.3-JA2_DN50_stripped_peptides.fas) as fasta_file:  # Will close handle cleanly
    identifiers = []
    lengths = []
    aa = []
    for seq_record in SeqIO.parse(fasta_file, 'fasta'):  # (generator)
        identifiers.append(seq_record.id)
        lengths.append(len(seq_record.seq))
        aa.count_amino_acids(seq_record.seq)
        
        
#converting lists to pandas Series    
s1 = Series(identifiers, name='ID')
s2 = Series(lengths, name='length')
s3 = Series(aa, name='AAs')

#Gathering Series into a pandas DataFrame and rename index as ID column
idseq = DataFrame(dict(ID=s1, length=s2, AAs=s3)).set_index(['ID'])

idseq.head()
```
