```python
# LIBRARIES
#import pandas library for working with tabular data
import os
os.getcwd()
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#from matplotlib import pyplot
import seaborn as sns
from scipy.stats import kde
#import regular expresson (regex)
import re
#check pandas version
pd.__version__
```

```python
# read the CSVs of each replicate into a datadrame we name 'peaksdb' using the pandas read_csv function
peaksdb1 = pd.read_csv("/home/millieginty/Documents/git-repos/2017-etnp/data/pro2020/RAL4_95_MED2_trypsin/PeaksDB-PTMopt/RAL_95_MED2_trypsin_1_PTMopt_DB20/MED2_1_PTMopt_DB20_peptide.csv")
peaksdb2 = pd.read_csv("/home/millieginty/Documents/git-repos/2017-etnp/data/pro2020/RAL4_95_MED2_trypsin/PeaksDB-PTMopt/RAL_95_MED2_trypsin_2_PTMopt_DB20/MED2_2_PTMopt_DB20_peptide.csv")

# rename the columns, since they contain injection info that makes concatenating difficult
column_name = ['peptide', '-10lgP', 'Mass', 'Length', 'ppm', 'm/z', 'RT',
       'Area', 'Fraction', 'Scan', 'Source File', '#Spec_tot',
       '#Spec', 'Accession', 'PTM', 'AScore']
peaksdb1.columns=column_name
peaksdb2.columns=column_name

print(peaksdb1.shape)
print(peaksdb2.shape)

print(peaksdb1.columns)
print(peaksdb2.columns) 

frames = [peaksdb1, peaksdb2]

# append dataframes
peaksdbdup = peaksdb1.append(peaksdb2)

# remove redundant rows
peaksdb = pd.DataFrame.drop_duplicates(peaksdbdup)

print("# redundant peaksdb peptides in combined dataframe", len(peaksdbdup))
print("# nonredundant peaksdb peptides in combined dataframe", len(peaksdb))

peaksdb.head()
```

The peptide column has the modifications (e.g., 57.02 Da for carbamidomethylation of cysteine). We want to make new columns with all that information and make a column with only the 'stripped' peptide sequence that's just amino acids - this we can then align against other sequences, for example.

Modified residues were allowed for:

    fixed carbamodimethylation of cysteine 57.021464 C
    varialbe oxidation of methionine: 15.9949 M
    variable deamidation of asparagine, glumatine: 0.984016 NQ
    variable iron cation on lysine: 53.919289 K
    variable methylation of lysine and arginine: 14.015650 KR

We'll then write this manipulated dataframe to a new file.

```python
# use a count function to enumerate the # of A's (alanines) in each peptide
peaksdb['A'] = peaksdb['peptide'].str.count("A")

# use a count function to enumerate the # of C's (cysteines) in each peptide
peaksdb['C'] = peaksdb['peptide'].str.count("C")

# use a count function to enumerate the # of D's (aspartic acids) in each peptide
peaksdb['D'] = peaksdb['peptide'].str.count("D")

# use a count function to enumerate the # of E's (glutamic acids) in each peptide
peaksdb['E'] = peaksdb['peptide'].str.count("E")

# use a count function to enumerate the # of F's (phenylalanines) in each peptide
peaksdb['F'] = peaksdb['peptide'].str.count("F")

# use a count function to enumerate the # of G's (glycines) in each peptide
peaksdb['G'] = peaksdb['peptide'].str.count("G")

# use a count function to enumerate the # of H's (histidines) in each peptide
peaksdb['H'] = peaksdb['peptide'].str.count("H")

# use a count function to enumerate the # of I's (isoleucines) in each peptide
# in peaksdb output, there will be no isoleucines (they're lumped in with leucines)
peaksdb['I'] = peaksdb['peptide'].str.count("I")

# use a count function to enumerate the # of K's (lysines) in each peptide
peaksdb['K'] = peaksdb['peptide'].str.count("K")

# use a count function to enumerate the # of L's (leucines) in each peptide
peaksdb['L'] = peaksdb['peptide'].str.count("L")

# use a count function to enumerate the # of M's (methionines) in each peptide
peaksdb['M'] = peaksdb['peptide'].str.count("M")

# use a count function to enumerate the # of N's (asparagines) in each peptide
peaksdb['N'] = peaksdb['peptide'].str.count("N")

# use a count function to enumerate the # of P's (prolines) in each peptide
peaksdb['P'] = peaksdb['peptide'].str.count("P")

# use a count function to enumerate the # of Q's (glutamines) in each peptide
peaksdb['Q'] = peaksdb['peptide'].str.count("Q")

# use a count function to enumerate the # of R's (arginines) in each peptide
peaksdb['R'] = peaksdb['peptide'].str.count("R")

# use a count function to enumerate the # of S's (serines) in each peptide
peaksdb['S'] = peaksdb['peptide'].str.count("S")

# use a count function to enumerate the # of T's (threonines) in each peptide
peaksdb['T'] = peaksdb['peptide'].str.count("T")

# use a count function to enumerate the # of V's (valines) in each peptide
peaksdb['V'] = peaksdb['peptide'].str.count("V")

# use a count function to enumerate the # of W's (tryptophans) in each peptide
peaksdb['W'] = peaksdb['peptide'].str.count("W")

# use a count function to enumerate the # of Y's (tyrosines) in each peptide
peaksdb['Y'] = peaksdb['peptide'].str.count("Y")

# use a count function to enumerate the # of carbamidomethylated C's in each peptide
peaksdb['c-carb'] = peaksdb['peptide'].str.count("57.02")

# use a count function to enumerate the # of oxidized M's in each peptide
peaksdb['m-oxid'] = peaksdb['peptide'].apply(lambda x: x.count('M(+15.99)'))

# use a lamba function to enumerate the # of deamidated N's in each peptide
# used to use peaksdb['n-deam'] = peaksdb['peptide'].str.count("N\(+.98") but that didn't work with the 'N'

peaksdb['n-deam'] = peaksdb['peptide'].apply(lambda x: x.count('N(+.98)'))

# use a count function to enumerate the # of deamidated Q's in each peptide
peaksdb['q-deam'] = peaksdb['peptide'].apply(lambda x: x.count('Q(+.98)'))

# use a count function to enumerate the # of iron adducted K's in each peptide
peaksdb['k-iron'] = peaksdb['peptide'].str.count("53.92")

# use a count function to enumerate the # of methylated K's in each peptide
peaksdb['k-meth'] = peaksdb['peptide'].apply(lambda x: x.count('K(+14.02)'))

# use a count function to enumerate the # of methylated R's in each peptide
peaksdb['r-meth'] = peaksdb['peptide'].apply(lambda x: x.count('R(+14.02)'))

# create a column with 'stripped' peptide sequences using strip
peaksdb['stripped peptide'] = peaksdb['peptide'].str.replace(r"\(.*\)","")

# add a column with the stripped peptide length (number of AAs)
peaksdb['stripped length'] = peaksdb['stripped peptide'].apply(len)

# total the number of modifications in sequence
peaksdb['ptm-total'] = peaksdb['c-carb'] + peaksdb['m-oxid'] + peaksdb['n-deam'] + peaksdb['q-deam'] + peaksdb['k-iron'] + peaksdb['k-meth'] + peaksdb['r-meth']

# calculate the NAAF numerator for later NAAF normalization
peaksdb['NAAF num.'] = peaksdb['Area'] / peaksdb['stripped length']

# write modified dataframe to new txt file, plus 'unfiltered'
peaksdb.to_csv("/home/millieginty/Documents/git-repos/2017-etnp/data/pro2020/RAL4_95_MED2_trypsin/PeaksDB-PTMopt/RAL4_MED2_trypsin_1_PTMopt_PeaksDB_unfiltered.csv")

# check out the results
peaksdb.head()
```

## NAAF correction and exporting files with AA and PTM totals:

```python
# made a new dataframe that contains the sums of certain columns in the stripped peptide dataframe above

index = ['sample total']

data = {'A': peaksdb['A'].sum(),
        'C': peaksdb['C'].sum(),
        'D': peaksdb['D'].sum(),
        'E': peaksdb['E'].sum(),
        'F': peaksdb['F'].sum(),
        'G': peaksdb['G'].sum(),
        'H': peaksdb['H'].sum(),
        'I': peaksdb['I'].sum(),
        'K': peaksdb['K'].sum(),
        'L': peaksdb['L'].sum(),
        'M': peaksdb['M'].sum(),
        'N': peaksdb['N'].sum(),
        'P': peaksdb['P'].sum(),
        'Q': peaksdb['Q'].sum(),
        'R': peaksdb['R'].sum(),
        'S': peaksdb['S'].sum(),
        'T': peaksdb['T'].sum(),
        'V': peaksdb['V'].sum(),
        'W': peaksdb['W'].sum(),
        'Y': peaksdb['Y'].sum(),
        'c-carb': peaksdb['c-carb'].sum(),
        'm-oxid': peaksdb['m-oxid'].sum(),
        'n-deam': peaksdb['n-deam'].sum(),
        'q-deam': peaksdb['q-deam'].sum(),
        'k-iron': peaksdb['k-iron'].sum(),
        'k-meth': peaksdb['k-meth'].sum(),
        'r-meth': peaksdb['r-meth'].sum(),
        'Total area': peaksdb['Area'].sum(),
        'Total length': peaksdb['stripped length'].sum()
       }

totalpeaksdb = pd.DataFrame(data, columns=['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', \
                                           'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', \
                                           'W', 'Y', 'c-carb', 'm-oxid', 'n-deam', \
                                           'q-deam', 'k-iron', 'k-meth', 'r-meth', \
                                          'Total area', 'Total length'], index=index)

# calculate percentage of C's with carb (should be 1.0)
totalpeaksdb['% C w/ carb.'] = totalpeaksdb['c-carb'] / totalpeaksdb['C'] 

# calculate percentage of M's that are oxidized
totalpeaksdb['% M w/ oxid'] = totalpeaksdb['m-oxid'] / totalpeaksdb['M'] 

# calculate percentage of N's that are deamidated
totalpeaksdb['% N w/ deam'] = totalpeaksdb['n-deam'] / totalpeaksdb['N'] 

# calculate percentage of Q's that are deamidated
totalpeaksdb['% Q w/ deam'] = totalpeaksdb['q-deam'] / totalpeaksdb['Q'] 

# calculate percentage of K's that are hydroxylated
totalpeaksdb['% K w/ iron'] = totalpeaksdb['k-iron'] / totalpeaksdb['K'] 

# calculate percentage of K's that are methylated
totalpeaksdb['% K w/ meth'] = totalpeaksdb['k-meth'] / totalpeaksdb['K'] 

# calculate percentage of R's that are methylated
totalpeaksdb['% R w/ meth'] = totalpeaksdb['r-meth'] / totalpeaksdb['R'] 

# calculate NAAF denominator for all peptides in dataset i
totalpeaksdb['NAAF denom.'] = totalpeaksdb['Total area'] / totalpeaksdb['Total length']

# write modified dataframe to new txt file, same name + totals
totalpeaksdb.to_csv("/home/millieginty/Documents/git-repos/2017-etnp/data/pro2020/RAL4_95_MED2_trypsin/PeaksDB-PTMopt/RAL95_MED2_trypsin_combine_PeaksDB_totals.csv")

totalpeaksdb.head()
```

```python
# use the calculated NAAF factor (from above) to caluclate the NAAF 
# NAAF: normalized normalized area abundance factor

NAAF20 = 3.689440e+06

# use NAAF >XCorr 3 to get NAAF
peaksdb['NAAF factor'] = (peaksdb['NAAF num.'])/NAAF20

# make a dataframe that contains only what we need: sequences, AAs, PTMs
peaksdb_AA = peaksdb[['stripped peptide', 'NAAF factor', 'A', 'C', 'D', 'E', 'F', 'G', 'H', 'K', 'I', 'L', \
                                'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y', 'c-carb', 'm-oxid', 'n-deam', \
                                'q-deam', 'k-iron', 'k-meth', 'r-meth']].copy()

# multiply the NAAF20 factor by the AAs to normalize its abundance by peak area and peptide length

peaksdb_AA['A-NAAF20'] = peaksdb_AA['A'] * peaksdb['NAAF factor']
peaksdb_AA['C-NAAF20'] = peaksdb_AA['C'] * peaksdb['NAAF factor']
peaksdb_AA['D-NAAF20'] = peaksdb_AA['D'] * peaksdb['NAAF factor']
peaksdb_AA['E-NAAF20'] = peaksdb_AA['E'] * peaksdb['NAAF factor']
peaksdb_AA['F-NAAF20'] = peaksdb_AA['F'] * peaksdb['NAAF factor']
peaksdb_AA['G-NAAF20'] = peaksdb_AA['G'] * peaksdb['NAAF factor']
peaksdb_AA['H-NAAF20'] = peaksdb_AA['H'] * peaksdb['NAAF factor']
peaksdb_AA['K-NAAF20'] = peaksdb_AA['K'] * peaksdb['NAAF factor']
peaksdb_AA['I-NAAF20'] = peaksdb_AA['I'] * peaksdb['NAAF factor']
peaksdb_AA['L-NAAF20'] = peaksdb_AA['L'] * peaksdb['NAAF factor']
peaksdb_AA['M-NAAF20'] = peaksdb_AA['M'] * peaksdb['NAAF factor']
peaksdb_AA['N-NAAF20'] = peaksdb_AA['N'] * peaksdb['NAAF factor']
peaksdb_AA['P-NAAF20'] = peaksdb_AA['P'] * peaksdb['NAAF factor']
peaksdb_AA['Q-NAAF20'] = peaksdb_AA['Q'] * peaksdb['NAAF factor']
peaksdb_AA['R-NAAF20'] = peaksdb_AA['R'] * peaksdb['NAAF factor']
peaksdb_AA['S-NAAF20'] = peaksdb_AA['S'] * peaksdb['NAAF factor']
peaksdb_AA['T-NAAF20'] = peaksdb_AA['T'] * peaksdb['NAAF factor']
peaksdb_AA['V-NAAF20'] = peaksdb_AA['V'] * peaksdb['NAAF factor']
peaksdb_AA['W-NAAF20'] = peaksdb_AA['W'] * peaksdb['NAAF factor']
peaksdb_AA['Y-NAAF20'] = peaksdb_AA['Y'] * peaksdb['NAAF factor']

# multiply the NAAF20 factor by the PTMs normalize its abundance by peak area and peptide length

peaksdb_AA['ccarb-NAAF20'] = peaksdb_AA['c-carb'] * peaksdb_AA['NAAF factor']
peaksdb_AA['moxid-NAAF20'] = peaksdb_AA['m-oxid'] * peaksdb_AA['NAAF factor']
peaksdb_AA['ndeam-NAAF20'] = peaksdb_AA['n-deam'] * peaksdb_AA['NAAF factor']
peaksdb_AA['qdeam-NAAF20'] = peaksdb_AA['q-deam'] * peaksdb_AA['NAAF factor']
peaksdb_AA['kiron-NAAF20'] = peaksdb_AA['k-iron'] * peaksdb_AA['NAAF factor']
peaksdb_AA['kmeth-NAAF20'] = peaksdb_AA['k-meth'] * peaksdb_AA['NAAF factor']
peaksdb_AA['rmeth-NAAF20'] = peaksdb_AA['r-meth'] * peaksdb_AA['NAAF factor']

# write the dataframe to a new csv
peaksdb_AA.to_csv("/home/millieginty/Documents/git-repos/2017-etnp/data/pro2020/RAL4_95_MED2_trypsin/PeaksDB-PTMopt/RAL4_MED2_combine_PeaksDB_AA_NAAF.csv")

peaksdb_AA.head()
```

```python
# made a new dataframe that contains the sums of certain columns in the stripped peptide dataframe above

index = ['sample total']

data = {'A': peaksdbpsm['A'].sum(),
        'C': peaksdbpsm['C'].sum(),
        'D': peaksdbpsm['D'].sum(),
        'E': peaksdbpsm['E'].sum(),
        'F': peaksdbpsm['F'].sum(),
        'G': peaksdbpsm['G'].sum(),
        'H': peaksdbpsm['H'].sum(),
        'I': peaksdbpsm['I'].sum(),
        'K': peaksdbpsm['K'].sum(),
        'L': peaksdbpsm['L'].sum(),
        'M': peaksdbpsm['M'].sum(),
        'N': peaksdbpsm['N'].sum(),
        'P': peaksdbpsm['P'].sum(),
        'Q': peaksdbpsm['Q'].sum(),
        'R': peaksdbpsm['R'].sum(),
        'S': peaksdbpsm['S'].sum(),
        'T': peaksdbpsm['T'].sum(),
        'V': peaksdbpsm['V'].sum(),
        'W': peaksdbpsm['W'].sum(),
        'Y': peaksdbpsm['Y'].sum(),
        'c-carb': peaksdbpsm['c-carb'].sum(),
        'm-oxid': peaksdbpsm['m-oxid'].sum(),
        'n-deam': peaksdbpsm['n-deam'].sum(),
        'q-deam': peaksdbpsm['q-deam'].sum(),
        'k-iron': peaksdbpsm['k-iron'].sum(),
        'k-meth': peaksdbpsm['k-meth'].sum(),
        'r-meth': peaksdbpsm['r-meth'].sum(),
        'Total area': peaksdbpsm['Area'].sum(),
        'Total length': peaksdbpsm['stripped length'].sum()
       }

totalpeaksdbpsm = pd.DataFrame(data, columns=['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', \
                                           'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', \
                                           'W', 'Y', 'c-carb', 'm-oxid', 'n-deam', \
                                           'q-deam', 'k-iron', 'k-meth', 'r-meth', \
                                          'Total area', 'Total length'], index=index)

# calculate percentage of C's with carb (should be 1.0)
totalpeaksdbpsm['% C w/ carb.'] = totalpeaksdbpsm['c-carb'] / totalpeaksdbpsm['C'] 

# calculate percentage of M's that are oxidized
totalpeaksdbpsm['% M w/ oxid'] = totalpeaksdbpsm['m-oxid'] / totalpeaksdbpsm['M'] 

# calculate percentage of N's that are deamidated
totalpeaksdbpsm['% N w/ deam'] = totalpeaksdbpsm['n-deam'] / totalpeaksdbpsm['N'] 

# calculate percentage of Q's that are deamidated
totalpeaksdbpsm['% Q w/ deam'] = totalpeaksdbpsm['q-deam'] / totalpeaksdbpsm['Q'] 

# calculate percentage of K's that are hydroxylated
totalpeaksdbpsm['% K w/ iron'] = totalpeaksdbpsm['k-iron'] / totalpeaksdbpsm['K'] 

# calculate percentage of K's that are methylated
totalpeaksdbpsm['% K w/ meth'] = totalpeaksdbpsm['k-meth'] / totalpeaksdbpsm['K'] 

# calculate percentage of R's that are methylated
totalpeaksdbpsm['% R w/ meth'] = totalpeaksdbpsm['r-meth'] / totalpeaksdbpsm['R'] 

# calculate NAAF denominator for all peptides in dataset i
totalpeaksdbpsm['NAAF denom.'] = totalpeaksdbpsm['Total area'] / totalpeaksdbpsm['Total length']

# write modified dataframe to new txt file, same name + totals
totalpeaksdbpsm.to_csv("/home/millieginty/Documents/git-repos/2017-etnp/data/pro2020/RAL4_95_MED2_trypsin/PeaksDB-PTMopt/RAL95_MED2_trypsin_combine_PeaksDBpsm_totals.csv")

totalpeaksdbpsm.head()
```
