```python
### Sum NAAF numerators and get NAAF corrected taxa % 

# Get the NAAF numerator totals for phylum-level ID'd peptides (NAAF num = Area/length)

pdb_NAAFsum = pdb273['NAAF num.'].sum()

# PeaksDB
# sum the NAAF numerators for Cyanobacteria peptides

pdbNAAF_273_cyano = pdb273[pdb273['phylum'].str.contains('Cyanobacteria')]
pdbNAAF_273_fungi = pdb273[pdb273['phylum'].str.contains('Ascomycota|Basidiomycota|Chytridiomycota|Mucoromycota|Zoopagomycota')]
pdbNAAF_273_hetb = pdb273[pdb273['phylum'].str.contains('Actinobacteria|Bacteroidetes|Firmicutes|Proteobacteria|Candidatus Marinimicrobia')]
pdbNAAF_273_nitro = pdb273[pdb273['phylum'].str.contains('Nitrospinae')]

pdbNAAF273_sum_cyano = pdbNAAF_273_cyano['NAAF num.'].sum()
pdbNAAF273_sum_fungi = pdbNAAF_273_fungi['NAAF num.'].sum()
pdbNAAF273_sum_hetb = pdbNAAF_273_hetb['NAAF num.'].sum()
pdbNAAF273_sum_nitro = pdbNAAF_273_nitro['NAAF num.'].sum()

print("% pdb Cyano = ", (pdbNAAF273_sum_cyano / pdb_NAAFsum)*100)
print("% pdb Fungi = ", (pdbNAAF273_sum_fungi / pdb_NAAFsum)*100)
print("% pdb Het. Bacteria = ", (pdbNAAF273_sum_hetb / pdb_NAAFsum)*100)
print("% pdb Nitrospina = ", (pdbNAAF273_sum_nitro / pdb_NAAFsum)*100)
print("% pdb covered by these catergories = ", ((pdbNAAF273_sum_cyano + pdbNAAF273_sum_fungi + \
                                             pdbNAAF273_sum_hetb + pdbNAAF273_sum_nitro)/pdb_NAAFsum)*100)
```

```python
### Sum NAAF numerators and get NAAF corrected taxa % 

# Get the NAAF numerator totals for phylum-level ID'd peptides (NAAF num = Area/length)

dn80_NAAFsum = dn80273['NAAF num.'].sum()

# PeaksDB
# sum the NAAF numerators for Cyanobacteria peptides

dn80NAAF_273_cyano = dn80273[dn80273['phylum'].str.contains('Cyanobacteria')]
dn80NAAF_273_fungi = dn80273[dn80273['phylum'].str.contains('Ascomycota|Basidiomycota|Chytridiomycota|Mucoromycota|Zoopagomycota')]
dn80NAAF_273_hetb = dn80273[dn80273['phylum'].str.contains('Actinobacteria|Bacteroidetes|Firmicutes|Proteobacteria|Candidatus Marinimicrobia')]
dn80NAAF_273_nitro = dn80273[dn80273['phylum'].str.contains('Nitrospinae')]

dn80NAAF273_sum_cyano = dn80NAAF_273_cyano['NAAF num.'].sum()
dn80NAAF273_sum_fungi = dn80NAAF_273_fungi['NAAF num.'].sum()
dn80NAAF273_sum_hetb = dn80NAAF_273_hetb['NAAF num.'].sum()
dn80NAAF273_sum_nitro = dn80NAAF_273_nitro['NAAF num.'].sum()

print("% dn80 Cyano = ", (dn80NAAF273_sum_cyano / dn80_NAAFsum)*100)
print("% dn80 Fungi = ", (dn80NAAF273_sum_fungi / dn80_NAAFsum)*100)
print("% dn80 Het. Bacteria = ", (dn80NAAF273_sum_hetb / dn80_NAAFsum)*100)
print("% dn80 Nitrospina = ", (dn80NAAF273_sum_nitro / dn80_NAAFsum)*100)
print("% dn80 covered by these catergories = ", ((dn80NAAF273_sum_cyano + dn80NAAF273_sum_fungi + \
                                             dn80NAAF273_sum_hetb + dn80NAAF273_sum_nitro)/dn80_NAAFsum)*100)
```

```python
# calculate a weighted average (based on # of peptides) of taxonomic peptide contribution
# of all method (Comet, PeaksDB, de novo)

#comet_peps = len(comet273)
pdb_peps = len(pdb273)
dn_peps = len(dn80273)

ave_273_cyano = ((pdb_peps*(pdbNAAF273_sum_cyano / pdb_NAAFsum)) + \
                (dn_peps*(dn80NAAF273_sum_cyano / dn80_NAAFsum))) / (pdb_peps + dn_peps)


ave_273_fungi = ((pdb_peps*(pdbNAAF273_sum_fungi / pdb_NAAFsum)) + \
                (dn_peps*(dn80NAAF273_sum_fungi / dn80_NAAFsum))) / (pdb_peps + dn_peps)

ave_273_hetb = ((pdb_peps*(pdbNAAF273_sum_hetb / pdb_NAAFsum)) + \
                (dn_peps*(dn80NAAF273_sum_hetb / dn80_NAAFsum))) / (pdb_peps + dn_peps)


ave_273_nitro = ((pdb_peps*(pdbNAAF273_sum_nitro / pdb_NAAFsum)) + \
                (dn_peps*(dn80NAAF273_sum_nitro / dn80_NAAFsum))) / (pdb_peps + dn_peps)


print("Weighted ave. contribution of Cyano peptides =", ave_273_cyano*100)
print("Weighted ave. contribution of Fungi peptides =", ave_273_fungi*100)
print("Weighted ave. contribution of heterotrophic bac. peptides =", ave_273_hetb*100)
print("Weighted ave. contribution of Nitrospina peptides =", ave_273_nitro*100)
```
