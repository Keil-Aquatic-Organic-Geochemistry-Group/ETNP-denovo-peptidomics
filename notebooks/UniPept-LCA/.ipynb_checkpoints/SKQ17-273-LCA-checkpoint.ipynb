```python
dn80_lca273 = pd.read_csv('De-novo/273-PTMopt-DN80-lca-tax.csv')

# drop the peptide that aren't specific to the phylum level
dn80_phy273 = dn80_lca273[dn80_lca273.phylum.notnull()]

# drop everything else
dn80_phy273 = dn80_phy273[['peptide', 'phylum']].copy()

print('Peptides specific to the Phylum level:', len(dn80_phy273))
phylum = len(dn80_phy273)

#Cyanobacteria
print('Peptide specific to Cyanobacteria:', len(dn80_phy273[dn80_phy273['phylum'].str.contains('Cyanobacteria')]))

Cyanobacteria = len(dn80_phy273[dn80_phy273['phylum'].str.contains('Cyanobacteria')])

#Fungi and fungi-like
print('Peptide specific to Ascomycota:', len(dn80_phy273[dn80_phy273['phylum'].str.contains('Ascomycota')]))
print('Peptide specific to Basidiomycota:', len(dn80_phy273[dn80_phy273['phylum'].str.contains('Basidiomycota')]))
print('Peptide specific to Chytridiomycota:', len(dn80_phy273[dn80_phy273['phylum'].str.contains('Chytridiomycota')]))
print('Peptide specific to Mucoromycota:', len(dn80_phy273[dn80_phy273['phylum'].str.contains('Mucoromycota')]))
print('Peptide specific to Zoopagomycota:', len(dn80_phy273[dn80_phy273['phylum'].str.contains('Zoopagomycota')]))

Ascomycota = len(dn80_phy273[dn80_phy273['phylum'].str.contains('Ascomycota')])
Basidiomycota = len(dn80_phy273[dn80_phy273['phylum'].str.contains('Basidiomycota')])
Chytridiomycota= len(dn80_phy273[dn80_phy273['phylum'].str.contains('Chytridiomycota')])
Mucoromycota = len(dn80_phy273[dn80_phy273['phylum'].str.contains('Mucoromycota')])
Zoopagomycota = len(dn80_phy273[dn80_phy273['phylum'].str.contains('Zoopagomycota')])

#Hetertrophic bacteria
print('Peptide specific to Actinobacteria:', len(dn80_phy273[dn80_phy273['phylum'].str.contains('Actinobacteria')]))
print('Peptide specific to Bacteroidetes:', len(dn80_phy273[dn80_phy273['phylum'].str.contains('Bacteroidetes')]))
print('Peptide specific to Firmicutes:', len(dn80_phy273[dn80_phy273['phylum'].str.contains('Firmicutes')]))
print('Peptide specific to Proteobacteria:', len(dn80_phy273[dn80_phy273['phylum'].str.contains('Proteobacteria')]))
print('Peptide specific to Candidatus Marinimicrobia:', len(dn80_phy273[dn80_phy273['phylum'].str.contains('Candidatus Marinimicrobia')]))

Actinobacteria = len(dn80_phy273[dn80_phy273['phylum'].str.contains('Actinobacteria')])
Bacteroidetes = len(dn80_phy273[dn80_phy273['phylum'].str.contains('Bacteroidetes')])
Firmicutes = len(dn80_phy273[dn80_phy273['phylum'].str.contains('Firmicutes')])
Proteobacteria = len(dn80_phy273[dn80_phy273['phylum'].str.contains('Proteobacteria')])
Candidatus_Marinimicrobia = len(dn80_phy273[dn80_phy273['phylum'].str.contains('Candidatus Marinimicrobia')])

#Nitrospina
print('Peptide specific to Nitrospirae:', len(dn80_phy273[dn80_phy273['phylum'].str.contains('Nitrospirae')]))

Nitrospirae = len(dn80_phy273[dn80_phy273['phylum'].str.contains('Nitrospirae')])

# make a dictionary of phylum output
phy_db80_273 = {"Cyanobacteria" : Cyanobacteria, "Ascomycota" : Ascomycota, "Basidiomycota" : Basidiomycota,\
                   "Chytridiomycota" : Chytridiomycota, "Mucoromycota" : Mucoromycota, \
                "Zoopagomycota" : Zoopagomycota, "Actinobacteria" : Actinobacteria, "Bacteroidetes" : Bacteroidetes, \
               "Firmicutes" : Firmicutes, "Proteobacteria" : Proteobacteria, "Nitrospirae" : Nitrospirae, \
              "Candidatus Marinimicrobia" : Candidatus_Marinimicrobia}

# make phylum specific dataframes
dn80_273_cyano = dn80_phy273[dn80_phy273['phylum'].str.contains('Cyanobacteria')]
dn80_273_fungi = dn80_phy273[dn80_phy273['phylum'].str.contains('Ascomycota|Basidiomycota|Chytridiomycota|Mucoromycota|Zoopagomycota')]
dn80_273_hetb = dn80_phy273[dn80_phy273['phylum'].str.contains('Actinobacteria|Bacteroidetes|Firmicutes|Proteobacteria|Candidatus Marinimicrobia')]
dn80_273_nitro = dn80_phy273[dn80_phy273['phylum'].str.contains('Nitrospirae')]

# save as a csv
dn80_phy273.to_csv("De-novo/Phylum-out/273/273-PTMopt-DN80-lca-phy.csv")
dn80_273_cyano.to_csv("De-novo/Phylum-out/273/273-PTMopt-DN80-lca-cyano.csv")
dn80_273_fungi.to_csv("De-novo/Phylum-out/273/273-PTMopt-DN80-lca-fungi.csv")
dn80_273_hetb.to_csv("De-novo/Phylum-out/273/273-PTMopt-DN80-lca-hetb.csv")
dn80_273_nitro.to_csv("De-novo/Phylum-out/273/273-PTMopt-DN80-lca-nitro.csv")

dn80_273_hetb.head()
```
